```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
#let's parse the dates for later
df = pd.read_csv('kc_house_data.csv', parse_dates = ['date'])
df.head(10)
```

```python
df.columns
```

ID - Unique ID for each home sold

date- Date of the home sale

price - Price of each home sold

bedrooms - Number of bedrooms

bathrooms - Number of bathrooms, where .5 accounts for a room with a toilet but no shower

sqft_living - Square footage of the apartments interior living space

sqft_lot - Square footage of the land space

floors - Number of floors

waterfront - A dummy variable for whether the apartment was overlooking the waterfront or not

view - An index from 0 to 4 of how good the view of the property was

condition - An index from 1 to 5 on the condition of the apartment,

grade - An index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of 
construction and design, and 11-13 have a high quality level of construction and design.

sqft_above - The square footage of the interior housing space that is above ground level

sqft_basement - The square footage of the interior housing space that is below ground level

yr_built - The year the house was initially built

yr_renovated - The year of the house’s last renovation

zipcode - What zipcode area the house is in

lat - Lattitude

long - Longitude

sqft_living15 - The square footage of interior housing living space for the nearest 15 neighbors

sqft_lot15 - The square footage of the land lots of the nearest 15 neighbors

```python
#lets remove the weird value
print(df[df.bedrooms >30])
df = df[df.bedrooms <30]
```

```python
feature_col = ['sqft_above', 'sqft_lot','sqft_lot15','sqft_living15','sqft_living','sqft_basement',
              'bedrooms','bathrooms','zipcode','floors','waterfront','view','grade',
              'month', 'year', 'long', 'lat', 'yr_built', 'yr_renovated','condition']

X=df[feature_col] #or   x = df.drop(['price', 'id', 'date'], axis = 1)
y = df.price #now its a series
```

```python
print(X.shape)
print(y.shape)
y.head()
```

```python
#split the data to train/test.

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.3,random_state=42)
```

```python
#Linnear regression
reg=LinearRegression()
reg.fit(X_train,y_train)#to train the model, we basiclly learn the coefficiends:MEKADMEI A FITEAURES
print(reg.intercept_)

```

```python
list(zip(feature_col,reg.coef_))
```

```python
predictions = reg.predict(X_test)
```

```python
#calculate evalution metrics
print('MAE:',metrics.mean_absolute_error(y_test, predictions))
print('MSE:',metrics.mean_squared_error(y_test, predictions))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('y_test')
plt.ylabel('PREDICTION')
#along x we have the test values, along y we have the predited valus
```

```python
predictions = reg.predict(X_train)
print('MAE:',metrics.mean_absolute_error(y_train, predictions))
print('MSE:',metrics.mean_squared_error(y_train, predictions))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_train, predictions)))
```

```python
plt.scatter(y_train,predictions)
plt.xlabel('y_train')
plt.ylabel('PREDICTION')
#along x we have the test values, along y we have the predited valus
```

```python

decision_tree = DecisionTreeRegressor()
decision_tree.fit(X_train, y_train)
predictions = decision_tree.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
#along x we have the test values, along y we have the predited valus
```

```python
print('MAE:',metrics.mean_absolute_error(y_test, predictions))
print('MSE:',metrics.mean_squared_error(y_test, predictions))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
